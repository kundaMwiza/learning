```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("output_timings.csv")
```

```python
df.loc[:, "mean_time"] = df.apply(lambda x: np.mean(x[3:]), axis = 1)
df.loc[:, "block_size"] = df.block_size.map(lambda x : "Naive_impl" if pd.isna(x) else x)
```

```python
filt_df = df[df.block_size != "Naive_impl"]
```

```python
filt_df
```

```python
p = sns.lineplot(data = df, x = "mat_dim", y = "mean_time", hue = "block_size")
p.set(ylabel = "mean_time(s)", title = "Matrix multiplication - blocked vs naive")

fig = p.get_figure()
fig.savefig("./timings.png")
```
